# 先从简单的开始，这个是用AC解决离散动作问题的代码

```python
import numpy as np
import tensorflow as tf
import gym
gpus = tf.config.experimental.list_physical_devices(device_type='GPU')
tf.config.experimental.set_memory_growth(gpus[0], True)
```

```python
# 超参数
MAX_EPISODE = 3000
DISPLAY_REWARD_THRESHOLD = 90  # renders environment if total episode reward is greater then this threshold
MAX_EP_STEPS = 1000   # maximum time step in one episode
RENDER = False  # rendering wastes time
GAMMA = 0.9     # reward discount in TD error
LR_A = 0.001    # learning rate for actor
LR_C = 0.01     # learning rate for critic

env = gym.make('CartPole-v0')
env.seed(1)  # reproducible
env = env.unwrapped

N_F = env.observation_space.shape[0]
N_A = env.action_space.n
```

```python
class Actor():
    def __init__(self,
                 n_features,
                 n_actions,
                 lr=0.001):
        
        #建立网络
        self.s = tf.keras.layers.Input(shape=(n_features,), name='state')
        x = tf.keras.layers.Dense(20, activation='relu', name='l1')(self.s)
        x = tf.keras.layers.Dense(n_actions, activation='softmax', name='acts_prob')(x)
        
        self.Actor_net = tf.keras.Model(inputs=self.s, outputs=x)
        
        #设置优化器
        self.a_optimizer = tf.keras.optimizers.Adam(learning_rate=lr)
        
    def learn(self, s, a, td_error):
        '''真实的参数都传进来'''
        s = np.expand_dims(s, axis=0)  #给s加一个维度，以便于可以输入网络
        
        with tf.GradientTape() as tape:

            acts_prob = self.Actor_net(s)
            print('动作概率为：',acts_prob)
            log_prob = tf.math.log(acts_prob[0, a])
            print('选择的动作对应的概率的对数：',log_prob)
            exp_v = tf.reduce_mean(log_prob * td_error)
            print('log_prob*td_error:',log_prob * td_error)
            print('v的期望,平均的log_prob*td_error：', exp_v)
            loss = -exp_v
            
        #计算梯度    
        gradients = tape.gradient(loss, self.Actor_net.trainable_variables)
        #优化
        self.a_optimizer.apply_gradients(zip(gradients, self.Actor_net.trainable_variables))
        
        return exp_v
    
    def choose_action(self, s):
        s = np.expand_dims(s, axis=0)  #给s加一个维度，以便于可以输入网络
        probs = self.Actor_net(s).numpy()
        act_choosed = np.random.choice(np.arange(probs.shape[1]), p=probs.ravel())
        return act_choosed
```

```python
class Critic():
    def __init__(self,
                 n_features,
                 lr=0.01):
        #建立网络
        self.s = tf.keras.layers.Input(shape=(n_features,), name='state')
        x = tf.keras.layers.Dense(20, activation='relu', name='l1')(self.s)
        x = tf.keras.layers.Dense(1, name='V')(x)
        
        self.Critic_net = tf.keras.Model(inputs=self.s, outputs=x)
        
        #设置优化器
        self.c_optimizer = tf.keras.optimizers.Adam(learning_rate=lr)
    
    def learn(self, s, r, s_):
        s = np.expand_dims(s, axis=0)
        s_ = np.expand_dims(s_, axis=0)
        
        with tf.GradientTape() as tape:
            v_ = self.Critic_net(s_)
            print('s_的价值v_',v_)
            v = self.Critic_net(s)
            print('s的价值v',v)
            td_error = r + GAMMA * v_ - v
            loss = tf.square(td_error)
            print('td_error的平方:',loss)
        
        #计算梯度
        gradients = tape.gradient(loss, self.Critic_net.trainable_variables)
        #优化
        self.c_optimizer.apply_gradients(zip(gradients, self.Critic_net.trainable_variables))
        
        return td_error
```

```python
actor = Actor(n_features=N_F, n_actions=N_A, lr=LR_A)
critic = Critic(n_features=N_F, lr=LR_C)     # we need a good teacher, so the teacher should learn faster than the actor
```

```python
for episod in range(MAX_EPISODE):
    s = env.reset()
    t = 0
    track_r = []
    
    while True:
        if RENDER: env.render()
        a = actor.choose_action(s)
        
        s_, r, done, info = env.step(a)
        
        if done:
            r = -20
        
        track_r.append(r)
        
        td_error = critic.learn(s, r, s_)
        actor.learn(s, a, td_error)
        
        s = s_
        t += 1
        
        if done or t >= MAX_EP_STEPS:
            ep_rs_sum = sum(track_r)
            
            if 'running_reward' not in globals():
                running_reward = ep_rs_sum
            else:
                running_reward = running_reward*0.95 + ep_rs_sum*0.05
            if running_reward > DISPLAY_REWARD_THRESHOLD: RENDER = True
            
            print('episod:', episod, 'reward:', int(running_reward))
            
            break
        
```
