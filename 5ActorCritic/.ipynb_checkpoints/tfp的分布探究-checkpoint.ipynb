```python
import numpy as np
import tensorflow as tf
import tensorflow_probability as tfp
import matplotlib.pyplot as plt
gpus = tf.config.experimental.list_physical_devices(device_type='GPU')
tf.config.experimental.set_memory_growth(gpus[0], True)
```

# 正态分布sample出来的样本究竟是什么！
# 就是一堆数，最小和最大附近的数少，中间位置的数多，
# 就好似[-10,-9,-8,-2.5,-2.4,-2.3,-1.9,-1.6,-1.4,-1.1,0,1.1,1.4,1.6,1.9,2.3,2.4,2.5,8,9,10]
# 接近0的地方数字产生的密集
# 就是相当于控制了步长，两端的数字步长长，越到中间步长越小，越密集

```python
mu = 1  #期望为1
sigma = 3  #标准差为3
num = 10000  #个数为10000

rand_data = np.random.normal(mu, sigma, num)  #符合高斯分布的样本，也就是横坐标X，越往两边的数越少，越往中间的数越多，也就是说sample的时候中间的更容易采到
print(rand_data)

# count统计的是某一bin出现的次数，如果density是True，那么count表示概率密度，这部分bin的概率就是count乘bin的长度，就是小矩形的面积
count, bins, ignored = plt.hist(rand_data, 50, density=True) # 分成50份,概率密度
print("counts are:",count)
print(count.shape)
print('bins are:',bins)
print(bins.shape)

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2)), linewidth=2, color='r')
plt.show()
```

```python
#每个矩形的宽为
width = (rand_data.max() - rand_data.min())/50
print(width)

#如果按我的理解count表示一个bin范围的概率密度，那么总的概率就是1
sum(count*width)
```

# 为什么bin有51个，而count只有50个？
# 因为bin51个两个两个之间空出了50个空，也就50个区间，count就是这50个区间上样本smaple出现的次数啊！

```python
def gassian(mu,sigma,x):
    '''返回正态概率密度'''
    f_x = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2))
    return f_x
```

# 检验tfp包里面的分布是怎么玩的

```python
mu = 1  #期望为1
sigma = 3  #标准差为3
num = 10000  #个数为10000

nm_dist = tfp.distributions.Normal(mu, sigma)
data = nm_dist.sample(num)

# sample出来的数据是打乱顺序排的，如果要使用他们画图，就应该把他们重新排列。
```

# 也就是说，choose动作的时候，按照这个正态分布sample，出来的值肯定在mu附近。因为mu在中间，采样到它的概率就大

# 不能对单个值说它的概率是多大，概率必须是要在区间上才有概率一说


# 但是这个log_prob到底是怎么算的呢？既然他只接受一个值，那么就代表它肯定不是概率，而是概率密度，我们知道，概率密度乘矩形宽度就是概率，如果把横轴分成超级小的等宽的小份，那么这一点的概率也就可以用概率密度乘小份的宽表示，既然是等长，那么这个点被选择的概率就和概率密度成正比了！

```python
log_prob = nm_dist.log_prob(1)  # normdist里面的log_prob就是这么算的！
# 所以说，log_prob(x)就是log(1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2)))

print(log_prob)  
print(np.exp(log_prob)) #得到的，就是x=1的概率密度

density_1 = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (1 - mu)**2 / (2 * sigma**2))
print(density_1)
```

```python
# 这个函数就是指数取对数变成减法啊！很简单！其实就是log(1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2)))变成了减法的形势
log_unnormalized = -0.5 * np.square(1/sigma-mu/sigma)  #分子
log_normalized = 0.5*np.log(2.*np.pi) + np.log(sigma)  #分母
log_unnormalized - log_normalized
print('未经归一化的log:',log_unnormalized)
print('归一化之后的log:',log_normalized)

print(log_unnormalized - log_normalized)
```

```python
nm_dist.entropy()
```

```python
entropy = (0.5 + log_normalized)  #分母+0.5
print(entropy)
```
